# Train Model

This code will be usede to train the a YOLO8 model on a custom data set.
The anotated data was previusly importet in the python file 'Data_import_VT2_mini.ipynb' 

```python
# Importing the required libraries
import os
from ultralytics import YOLO
```

```python
# Get the current working directory
HOME = os.getcwd()
print(f"Current working directory: {HOME}")
```

```python
# Load the YOLOv8 model
model = YOLO("yolov8s.pt")
```

```python
# Define the path to your dataset
dataset_path = f"{HOME}/VT2_mini_project-2"  #change path as needed 
```

```python
# Train the model
model.train(data=f"{dataset_path}/data.yaml", epochs=80,imgsz=640, patience=15, plots=True)
# epochs: number of epochs to train the model
# imgsz: image size for training 
# Larger image sizes can improve model accuracy by providing more detail, but they also increase memory usage and training time.
# Smaller image sizes reduce memory usage and training time but may result in lower accuracy.
# patience: number of epochs with no improvement after which training will be stopped
# plots: whether to plot training results
```
